```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# read and load the data to variables

data = pd.read_csv('data1.csv')
year_exp = data['YearsExperience'].values.reshape(-1,1)
salary = data['Salary'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(year_exp, salary, test_size = 1/3, random_state=42)
```

```python
#training the model

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#testing the model

y_pre = model.predict(X_test)
print(f" y_pre: {y_pre}")
print(f" y_test:{y_test}")
# measuring error

mse = mean_squared_error(y_test, y_pre)
print(f"mean squared error: {mse:.2f}")
rmse = pow(mse, 0.5)
print(f"root mean squared error: {rmse:.2f}")
r2 = r2_score(y_test, y_pre)
print(f"r2 score: {r2:.2f}")
```

```python
#ploting the data

plt.scatter(X_train,y_train, color = 'red', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Test Data')
plt.plot(X_test, y_pre, color = 'black')
plt.xlabel("experiance")
plt.ylabel("salary")
plt.show()
```
